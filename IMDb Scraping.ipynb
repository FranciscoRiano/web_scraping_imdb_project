Set-Up

```python
import requests
from bs4 import BeautifulSoup
```

```python
main_url = "https://www.imdb.com/search/name/?match_all=true&ref_=nv_cel_m"
```

```python
response = requests.get(main_url)
response
```

```python
html = response.content
html
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
with open('imdb_scraping.html','wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
divs = soup.find_all('div', {'class': 'lister-list'})
divs
```

```python
divsx = soup.find_all('div', {'class': 'lister-item mode-detail'})
divsx
```

```python
names = [div.find("h3")for div in divsx]
names
```

```python
artist_name = [name.find("a").string for name in names]
artist_name
```
